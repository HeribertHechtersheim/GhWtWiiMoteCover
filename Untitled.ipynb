```julia
using Images, FileIO, AxisArrays, ExifViewer
using Unitful: mm, inch
sdCardLen = 30.64mm
wallHei = 18mm
arr = [
    Dict(:imgPath=>"scans/IMG_0008.jpg", :len=>29.9mm, :padE=> 4.25mm),
    Dict(:imgPath=>"scans/IMG_0001.jpg", :len=>161.5mm, :padE=> 4.25mm),
    Dict(:imgPath=>"scans/IMG_0002.jpg", :len=>29.9mm, :padE=> 4.25mm),
    Dict(:imgPath=>"scans/IMG_0003.jpg", :len=>38.8mm, :padE=> 2.35mm),
    Dict(:imgPath=>"scans/IMG_0004.jpg", :len=>13.2mm, :padE=> 2.75mm),
    Dict(:imgPath=>"scans/IMG_0005.jpg", :len=>73.7mm, :padE=> 2.75mm),
    Dict(:imgPath=>"scans/IMG_0006.jpg", :len=>13.2mm, :padE=> 2.35mm),
    Dict(:imgPath=>"scans/IMG_0007.jpg", :len=>38.8mm, :padE=> 4.25mm),
    Dict(:imgPath=>"scans/IMG_0008.jpg", :len=>29.9mm, :padE=> 4.25mm)
]
images = []
for dict in arr
    imgPath = dict[:imgPath]
    len = dict[:len]
    padE = dict[:padE]
    tags = read_tags(imgPath)
    img = load(imgPath)
    strings = getindex.(Ref(tags), ["EXIF_TAG_X_RESOLUTION","EXIF_TAG_Y_RESOLUTION"])
    spacing = Tuple(map(s -> inch/parse(Float64, s) |> mm, strings))
    img = AxisArray(img, (:x, :y), spacing)
    img_size = size(img)
    img = img[sdCardLen..(Inf)mm, :]
    img = AxisArray(img, (:x, :y), spacing)    
    append!(images, [copy(img[0mm..len,0mm..wallHei])])
end
# croppedImg = img[50mm:100mm,:]
# summary(croppedImg)
# show(img)
# img_a = copy(img)
# zoom1 = img_a[5:20,80:150]
# zoom2 = @view img_a[2:10,2:10]
# fill!(zoom1, 0)
# IJulia.display(img_a)
```
